```python
# function that reads the data from the csv file, resizes the images and 
# returns train, test and validation sets according to the labels that already 
# exist in the dataset
def preprocess_data(file_path):
    data = pd.read_csv(file_path)

    train_set = data[(data.Usage == 'Training')]
    val_set = data[(data.Usage == 'PublicTest')]
    test_set = data[(data.Usage == 'PrivateTest')]

# This part with the Data Generators is in case we want more data to see if the 
# model trains better
# -----------------------------------------------------------------------------

    train_datagen = ImageDataGenerator(#rotation_range = 180,
                                            width_shift_range = 0.1,
                                            height_shift_range = 0.1,
                                            horizontal_flip = True
                                            )
                                            #zoom_range = 0.2)
                                            
    validation_datagen = ImageDataGenerator()

# -----------------------------------------------------------------------
    
    x_train = np.array(list(map(str.split, train_set.pixels)), np.float32)
    x_val = np.array(list(map(str.split, val_set.pixels)), np.float32)
    x_test = np.array(list(map(str.split, test_set.pixels)), np.float32)
    num_classes = 7
    # get labels and convert to categorical
    y_train = train_set.emotion
    y_train = to_categorical(y_train, num_classes)

    y_val = val_set.emotion
    y_val = to_categorical(y_val, num_classes)

    y_test = test_set.emotion
    y_test = to_categorical(y_test, num_classes)

    # reshape and normalize
    x_train = x_train.reshape((x_train.shape[0], 48, 48, 1)) / 255
    x_val = x_val.reshape((x_val.shape[0], 48, 48, 1)) / 255
    x_test = x_test.reshape((x_test.shape[0], 48, 48, 1)) / 255

    #------------------------------------------------
    train_flow = train_datagen.flow(x_train, y_train, batch_size=64) 
    val_flow = validation_datagen.flow(x_val, y_val, batch_size=64) 
    test_flow = validation_datagen.flow(x_test, y_test, batch_size=64)

    return x_train, x_val, x_test, y_train, y_val, y_test, train_flow, val_flow, test_flow

# function that implements and returns a model instance
def build_model():

  # vgg = VGG19(weights='imagenet', include_top=False)
  # for layer in vgg.layers:
  #     layer.trainable = False

  # input = Input(shape=(48, 48, 3))
  # conv = Conv2D(3, kernel_size=(3, 3), padding='same')(input)
  # vgg = vgg(conv)
  # x = Flatten()(vgg)
  # x = Dropout(0.3)(x)

  # pred = Dense(7, activation='softmax')(x)
  # model = Model(inputs=input, outputs=pred)

  model = Sequential()

  model.add(Conv2D(32, kernel_size=(3, 3), padding='same', activation='relu', input_shape=(48,48,1)))
  model.add(Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu'))
  # model.add(BatchNormalization())
  # model.add(Dropout(0.3))

  model.add(Conv2D(128, kernel_size=(3, 3), padding='same', activation='relu'))
  # model.add(BatchNormalization())
  model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
  model.add(Dropout(0.3))

  model.add(Conv2D(512, kernel_size=(3, 3), padding='same', activation='relu'))
  # model.add(BatchNormalization())
  model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
  model.add(Dropout(0.3))


  model.add(Conv2D(512, kernel_size=(3, 3), padding='same', activation='relu'))
  model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
  model.add(Dropout(0.3))

  model.add(Flatten())
  model.add(Dense(256, activation='relu'))

  model.add(Dense(512, activation='relu'))

  model.add(Dropout(0.3))
  model.add(Dense(7, activation='softmax'))

  return model

# plots the accuracy and loss history after fitting
def plot_history(history):

    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()

# plot some images from the sets. 
# the parameters howmany is used to define the range
def plot_images(x, howmany):
    for i in range(howmany):
        plt.figure(i)
        plt.imshow(x[i].reshape((48, 48)), interpolation='none', cmap='gray')
    plt.show()

label_names = ["angry", "disgust", "fear", "happy", "neutral", "sad", "surprise"]
# dictionary with the labels and the emotions
dict_label_index = {"0":"angry", "1":"disgust", "2":"fear", "3":"happy", "4":"neutral", "5":"sad", "6":"surprise"}
```

```python
def train_model(train_flow, val_flow, x_train, x_dev, x_test, y_test):
  model = build_model()
  model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.0009, decay=1e-6), metrics=['accuracy'])
  model.summary()
            
  batch_size=64
  hist = model.fit(train_flow, steps_per_epoch=len(x_train) / batch_size, epochs=20,
                  validation_data=val_flow,  
                  validation_steps=len(x_dev) / batch_size, 
                  callbacks = [EarlyStopping(patience=2, restore_best_weights=True), 
                                ReduceLROnPlateau(monitor='val_loss', 
                                factor=0.2, 
                                patience=6, 
                                verbose=1, 
                                min_delta=0.0001)])
  model.save("CNNModel")
  plot_history(hist)
  score = model.evaluate(x_test, y_test)
  print("Score on test set:", score)
```

# Use RISE to explain the predictions

```python
# RISE 
from skimage.transform import resize
from tqdm import tqdm

# generate masks, the values (48, 48) are according to the input size of the 
# neural net 
def generate_masks(N, s, p1):
    print(model.input_shape)
    cell_size = np.ceil(np.array((48, 48)) / s)
    up_size = (s + 1) * cell_size

    grid = np.random.rand(N, s, s) < p1
    grid = grid.astype('float32')

    masks = np.empty((N, 48, 48))

    for i in tqdm(range(N), desc='Generating masks'):
        
        # Random shifts
        x = np.random.randint(0, cell_size[0])
        y = np.random.randint(0, cell_size[1])
        # Linear upsampling and cropping
        masks[i, :, :] = resize(grid[i], up_size, order=1, mode='reflect',
                                anti_aliasing=False)[x:x + 48, y:y + 48] 
    print(masks.shape)
    masks = masks.reshape(-1, 48, 48, 1)
    return masks

batch_size = 100
# inp is image converted to array 
def explain(model, inp, masks):
    preds = []
    # Make sure multiplication is being done for correct axes
    masked = inp * masks
    for i in tqdm(range(0, N, batch_size), desc='Explaining'):
        preds.append(model.predict(masked[i:min(i+batch_size, N)]))
    preds = np.concatenate(preds)
    sal = preds.T.dot(masks.reshape(N, -1)).reshape(-1, 48, 48)
    sal = sal / N / p1
    return sal
```

```python
# Takes an image from the test set and plots the RISE explanation
def give_explanation_rise(img_num):

  x = np.expand_dims(x_test[img_num], axis=0) # get the img
  true_label = np.argmax(y_test[img_num]) # get the true label
  pred_label = np.argmax(model.predict(x)) # get the models prediction
  sal = explain(model, x, masks)
  print("True label: ", dict_label_index.get(str(true_label)))
  print("Predicted label:", dict_label_index.get(str(pred_label)))

  # show the image and the saliency map on top of it according to the true label
  plt.imshow(x.squeeze(), cmap='gray')
  plt.imshow(sal[true_label], cmap='jet', alpha=0.2)
  plt.colorbar()
  plt.show()
  return x, sal

# plot confusion matrix and classification report
def plot_cm(x_test, y_test):
  y_pred = np.argmax(model.predict(x_test), axis=1)
  y_test = np.argmax(y_test, axis=1)
  cm = confusion_matrix(y_test, y_pred)
  print(classification_report(y_test, y_pred, target_names=label_names))
  plt.figure(figsize=(8,8))
  plt.imshow(cm, interpolation='nearest')
  plt.colorbar()
  plt.title("Confusion matrix")
  tick_mark = np.arange(len(label_names))
  _ = plt.xticks(tick_mark, label_names, rotation=90)
  _ = plt.yticks(tick_mark, label_names)

plot_cm(x_test, y_test)
```
